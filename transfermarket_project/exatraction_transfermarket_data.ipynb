```python
import pandas as pd

# 1. Nuovo endpoint pubblico R2 ufficiale
BASE_URL = "https://pub-e682421888d945d684bcae8890b0ec20.r2.dev/data"

# 2. Header di sicurezza FONDAMENTALE per non farsi bloccare da Cloudflare (Evita l'errore 403)
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"}

print("Download e decompressione dei dati in corso (potrebbe volerci qualche secondo)...")

# 3. Pandas decomprime i file .gz, usando l'intestazione che "finge" di essere un browser
players = pd.read_csv(f"{BASE_URL}/players.csv.gz", storage_options=headers)
clubs = pd.read_csv(f"{BASE_URL}/clubs.csv.gz", storage_options=headers)

# 4. Filtra i club che appartengono alla Serie A (l'ID è 'IT1')
club_serie_a = clubs[clubs['domestic_competition_id'] == 'IT1']

# 5. Unisci i dati per ottenere solo i giocatori della Serie A
players_serie_a = players[players['current_club_id'].isin(club_serie_a['club_id'])]

# 6. Seleziona le colonne essenziali
df = players_serie_a[['first_name', 'last_name', 'name', 'last_season',
       'player_code', 'country_of_birth', 'date_of_birth', 'sub_position', 'position',
       'foot', 'height_in_cm', 'contract_expiration_date',
       'image_url', 'international_caps', 'international_goals',
       'url', 'current_club_name',
       'market_value_in_eur', 'highest_market_value_in_eur']].copy()

# Ordina per squadra in ordine alfabetico e per valore di mercato decrescente
df = df.sort_values(by=['current_club_name', 'market_value_in_eur'], ascending=[True, False])

print("--- Operazione completata! ---")
print(f"Estratti con successo {len(df)} giocatori della Serie A.")
```
